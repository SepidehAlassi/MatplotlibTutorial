# Matplotlib Tutorial



Matplotlib is a library for making 2D plots in Python, **matplotlib.pyplot** includes commands which are very similar to MATLAB commands, hence makes plotting with Python easier for beginners.  



## Basic Plotting

In order to see the resulting plots, matplotlib should be inlined in the first cell of IPython.

```python
%matplotlib inline
```

 To use the matplotlib functions matplotlib should be imported as **import matplotlib**,but by importing the **matplotlib.pyplot** one has access to all MATLB like commands of Pyplot. 

```python
import matplotlib
import matplotlib.pyplot as plt
```

**Plot()** command draws the plot, and is a versatile command which accepts arbitrary number of arguments.  In order to see the plots **show()** command should be used. One can easily give legend information to the plots by **title()**, **xlabel()**, **ylabel()**

```python
plt.plot([-1,0,1])
plt.title("simple plot")
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```

By giving a single list of numbers, matplotlib assumes the values in list as y-values and auto-generates a list of x-values with the same length (here 3).In Python indecies start with 0, that is why generated x values are [0,1,2].

### Controlling line properties and Formatting

The **axis()** command takes a list of [xmin, xmax, ymin, ymax] and specifies the visible limits of axes. The lines are drawn by default in "b-" format, anyother MATLAB line formatting can be specified. 

```python
plt.plot([1,2,3,4], [2,3,4,5], 'r--')
plt.axis([0,5,0,5])
```

There are other ways to specify the line formats, for example by using keywords such as **linewidth** etc as an argument of plot() function. **setp()** command can be used to to set style of multiple lines. 

```python
import numpy as np
X = np.linspace(-np.pi , np.pi, 256, endpoint=True)
C,S = np.cos(X), np.sin(X)

plt.plot(X,S, 'ro')
plt.plot(X,C, color="green", linewidth=2.0, linestyle="dotted")
```

```python
lines= plt.plot(X,S,X,C)
plt.setp(lines, color='blue',linewidth='2.0')
```

Texts can be added to the plots as a text placed by **text()** in arbitrary position. **annotate()** gives the possibility of annotating on the plot. In order to improve the scaling of the plots limits can be specified for the axes by **xlim()**, and **ylim()**. 

```python
plt.plot(X,S)
plt.text(1,1,'sin(X)',fontsize=12)
plt.plot(X,C, 'g--')
plt.annotate('cos(X)', xy=(-1,0.5), xytext=(-3,0.5), arrowprops=dict(facecolor='green', shrink=0.05))
plt.xlim(-4.0, 4,0)
plt.ylim(-1.2,1.2)

```
