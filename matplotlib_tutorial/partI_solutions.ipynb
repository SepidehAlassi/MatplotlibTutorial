Texts can be added to the plots as a text placed by **text()** in arbitrary position. **annotate()** gives the possibility of annotating points on the plot. In order to improve the scaling of the plots limits can be specified for the axes by **xlim()**, and **ylim()**. 

```python
plt.plot(X,C, 'b--')
t = 2*np.pi/3
plt.plot([t,t],[0,np.cos(t)], color ='blue', linewidth=1.5, linestyle="--")
# adding dot to the specific point
plt.scatter([t,],[np.cos(t),], 50, color ='blue')
#annotating a point in plot
plt.annotate(r'$\sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',
             xy=(t, np.sin(t)), xycoords='data',
             xytext=(-10, +30), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
# adding text to plot
plt.text(0,0,'sin(X)',fontsize=12, color='red')

plt.plot(X,S, 'r')
plt.plot([t,t],[0,np.sin(t)], color ='red', linewidth=1.5, linestyle="--")
plt.scatter([t,],[np.sin(t),], 50, color ='red')

plt.annotate(r'$\cos(\frac{2\pi}{3})=-\frac{1}{2}$',
             xy=(t, np.cos(t)), xycoords='data',
             xytext=(-80, -40), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
#text with arrow
plt.annotate('cos(X)',color='blue', xy=(-1,0.5), xytext=(-3,0.5), arrowprops=dict(facecolor='blue', shrink=0.05))
plt.xlim(-4.0, 4.0)
plt.ylim(-2.0,2.0)
```

### 1.3. Multiple plots and Figures
So far we have used implicit figure and axes creation. This is handy for fast plots. We can have more control over the display using figure, subplot, and axes explicitly. A figure in matplotlib means the whole window in the user interface. Within this figure there can be subplots. While subplot positions the plots in a regular grid, axes allows free placement within the figure. Both can be useful depending on your intention. We've already worked with figures and subplots without explicitly calling them.

#### 1.3.1. Figure
Figures are numbered starting from 1 as opposed to the normal Python way starting from 0. Various properties can be set for the figures, such as background color, title, etc. 

#### 1.3.2. Subplots
With subplot you can arrange plots in a regular grid. You need to specify the number of rows and columns and the number of the plot. For example in plt.subplot(212), 212  means the figure has 2 rows of subplots, arranged in 1 column, and this subplot is the 2nd plot. This arrangment of the subplots is important, by incorrect values the plots will be hidden. 


```python
fig1=plt.figure(1)
plt.plot(X,C, X,S)
#giving figure a title
fig1.suptitle('Figure1', fontsize=14, fontweight='bold')
#plotting a figure with subplots
fig2=plt.figure(2)
x2=plt.subplot(211, axisbg='black')
x2.plot(X,S, color='white')
x1=plt.subplot(212, axisbg='red')
x1.plot(X,C)
fig2.suptitle('Figure2', fontsize=14, fontstyle='italic')
```

#### 1.3.3. Sharing Axes
The subplots in one figure can share either x, or y axes or both of them.  

```python
t = np.arange(0.01, 5.0, 0.01)
s1 = np.sin(2*np.pi*t)
s2 = np.exp(-t)
s3 = np.sin(4*np.pi*t)

ax1 = plt.subplot(311)
plt.plot(t, s1)
plt.setp(ax1.get_xticklabels(), fontsize=6)

# share x only
ax2 = plt.subplot(312, sharex=ax1)
plt.plot(t, s2)
# make these tick labels invisible
plt.setp(ax2.get_xticklabels(), visible=False)

# share x and y
ax3 = plt.subplot(313, sharex=ax1, sharey=ax1)
plt.plot(t, s3)
plt.xlim(0.01, 5.0)
plt.show()
```

### 1.4. Spines
Spines are the lines connecting the axis tick marks and noting the boundaries of the data area. They can be placed at arbitrary positions and by default they are on the border of the axis. We'll change that since we want to have them in the middle. In the exmple below, we will move the left and buttom spines and make the top and right spines invisible. 

```python
fig, ax = plt.subplots()
ax.plot(x, y, 'k--')
# set ticks and tick labels
ax.set_xlim((0, 3*np.pi/2))
ax.set_xticks([0, np.pi/2, np.pi, 3*np.pi/2])
ax.set_xticklabels(['0', '$\pi$/2','$\pi$', '3$\pi$/2'])
ax.set_ylim((-1.5, 1.5))
ax.set_yticks([-1, 0, 1])
ax.spines['left'].set_bounds(-1, 1)
# Hide the right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
# Only show ticks on the left and bottom spines
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))
```

### 3.1.Bar Plots
Bar plots are quite often used to analyze statistics.**plt.bar()** is special for plotting bars. 
 

```python
import numpy as np
import matplotlib.pyplot as plt

n = 12
X = np.arange(n)
Y1 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
```

#### Aligning bars

```python
Y2 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)

#For a greater range of colors, you have two options. You can specify the color using an html hex string, as in:
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
#or you can pass an R , G , B tuple, where each of R , G , B are in the range [0,1].
plt.bar(X, -Y2, facecolor=(0.7,0.2,0.4), edgecolor='black')

for x,y in zip(X,Y1):
    plt.text(x+0.4, y+0.05, '%.2f' % y, ha='center', va= 'bottom')

plt.ylim(-1.25,1.25)
plt.show()
```

### 3.2. Histograms
in order to plot histograms, **hist()** command of pyplot can be used which also accepts different styling and formatting options as well as alpha value of the plot. **mlab** library of matplotlib contains numerical python functions written for compatability with MATLAB commands with the same names. Commands like **normpdf** of this library can be used to draw the best fit norm of the histogram. 

```python
import matplotlib.mlab as mlab
mu, sigma = 100, 15
x = mu + sigma*np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='green', alpha=0.5)

# add a 'best fit' line
y = mlab.normpdf( bins, mu, sigma)
l = plt.plot(bins, y, 'r--', linewidth=1)

plt.xlabel('Smarts')
plt.ylabel('Probability')
#title in latex style
plt.title(r'$\mathrm{Histogram\ of\ IQ:}\ \mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
#show the grid lines on the plot
plt.grid(True)

plt.show()
```

### 3.3. Scatter Plots
**scatter()** makes a scatter plot of x vs y, where x and y are sequence like objects of the same lengths.  
x, y : array of n numbers 
Input data:  
s : size in points^2. , scalar or array of n numbers, optional, default: 20  
 
c : color or sequence of color (optional) , c uses the colormap function to map single value to a color  
alpha: The alpha blending value, between 0 (transparent) and 1 (opaque)
    

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

plt.axes([0.025,0.025,0.95,0.95])
plt.scatter(X,Y, s=75,alpha=.5)

plt.xlim(-1.5,1.5), plt.xticks([])
plt.ylim(-1.5,1.5), plt.yticks([])

plt.show()
```

### 3.4. 3D-Plot
The library Axes3D is used for 3D plotting. By **Axes3D(fig)** the figure is set to be used for 3D plot.

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)

X = np.arange(-50, 50, 0.2)
Y = np.arange(-50, 50, 0.2)
# generate the mesh points
x, y = np.meshgrid(X, Y)
z= 1-abs(x+y)-abs(y-x) 

ax.plot_surface(x, y, z, cmap='hot')
plt.show()
```
